```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("IPCV.jpeg", 1)

if img is None:
    print("Error: Image could not be loaded.")
else:
    negative = 255 - img

    print("Pixel Grid of Original Image:\n")
    print(img)

    print("\nPixel Grid of Negative Image:\n")
    print(negative)

    plt.figure(figsize=(10,5))

    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(negative, cmap='gray')
    plt.title("Negative Image")
    plt.axis("off")

    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("IPCV.jpeg",0)

region = img[:8,:8]

plt.figure(figsize=(12,6))

for k in range(8):
    # Extract bit plane
    bit_plane = (region >> k) & 1

    print(f"Bit Plane {k} Grid Values:\n{bit_plane}\n")

    bit_img = bit_plane * 255

    full_plane = ((img >> k) & 1) * 255

    plt.subplot(2,4,k+1)
    plt.imshow(full_plane, cmap='gray')
    plt.title(f'Bit Plane {k}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# 4. Gray Level Slicing
gray_slice = np.zeros(img.shape, dtype=np.uint8)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    if 100 <= img[i][j] <= 180:
      gray_slice[i][j] = 255
```
