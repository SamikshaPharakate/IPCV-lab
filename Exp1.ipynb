```python
import cv2
from google.colab.patches import cv2_imshow # Import cv2_imshow

# 1. READ
img = cv2.imread('images.jpg', 1)

# Check if the image was loaded correctly
if img is None:
    print("Error: Could not find or open the image.")
else:
    height, width, _ = img.shape  # Unpack all three dimensions, ignoring the third
    print(f"Image loaded: {width}x{height} pixels")

    # 2. DISPLAY:
    cv2_imshow(img)

    # 3. WRITE: Save the image to a new file
    cv2.imwrite('saved_copy.png', img)
```

```python
import cv2
import matplotlib.pyplot as plt

# 1. Load the original color image (already in your code)
img = cv2.imread("images.jpg", 1)

if img is None:
    print("Error: Image could not be loaded.")
else:
    # 2. Convert to Grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 3. Display the results
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    # Convert BGR to RGB for correct display in Matplotlib
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original Color Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    # Display grayscale using the 'gray' colormap
    plt.imshow(gray_img, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis("off")

    plt.show()
```
